```python
from sklearn import datasets
import pandas as pd
```

```python
iris=datasets.load_iris()
```

```python
iris
```

```python
iris.keys()
```

```python
labels=pd.DataFrame(iris.target)
data=pd.DataFrame(iris.data)
```

```python
data
```

```python
labels.columns=['labels']
```

```python
data.columns=['Sepal length','Sepal width','Petal length','Petal width']
```

```python
labels
```

```python
data
```

# concat으로 추가

```python
data=pd.concat([data,labels],axis=1)
data
```

```python
feature=data[["Sepal length","Sepal width"]]
feature
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model=KMeans(n_clusters=3)# 숫자를 바꿔가며 클러스터 적절한것 알아내는것도 필요하다
model
```

# 데이터에 대한 훈련

```python
model.fit(feature)
```

# feature에 대한 예측 결과( 각각의 데이터가 어디 클러스터에 속한지를 확인)

```python
pred=pd.DataFrame(model.predict(feature))   #feature자리엔 원래 new data가 와야됨
pred
```

```python
pred.columns=["predict"]
```

```python
res=pd.concat([feature,pred],axis=1)
res
```

```python
data
```

```python
c_x=centers["Sepal length"]
c_y=centers["Sepal width"]
```

# 이 그래프는 반복이 끝나면 그래프가 그려짐

```python
wcss = [] # within cluster sum of square
for i in range(1,20):
    kmeans = KMeans(n_clusters=i,init='k-means++' ,max_iter=300,random_state=0)
    kmeans.fit(scaled_feature)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,20),wcss)
plt.title('THe Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel("WCSS")
plt.show()
```

- wcss = [] # within cluster sum of square
이 값이 작으면 작을 수록 많이 뭉쳐이다, 클러스터가 작은것
크면 클러스터 내부의 데이터가 멀리있다, 클러스터가 잘 안된것

- max_iter=300 : 300번이 넘도록 멈추지 않으면 강제 종료
